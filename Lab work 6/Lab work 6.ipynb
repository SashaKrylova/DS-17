```python
returns_df = pd.read_excel('/data/notebook_files/FOR_LAB_6.xlsx',sheet_name='Returns')
returns_df.head()
```

```python
grouped_by_order_date_orders_df = orders_df.groupby('Order Date')[['Order Date', 'Sales']].aggregate('sum').reset_index()
grouped_by_order_date_orders_df.head()
```

```python
f, ax = plt.subplots(figsize =(18,9))

ax.plot(grouped_by_order_date_orders_df['Order Date'], grouped_by_order_date_orders_df['Sales'])
ax.set_title('Time series')
ax.set_ylabel('Sales')
```

```python
grouped_by_ship_mode_orders_df = orders_df.groupby('Ship Mode')[['Ship Mode','Shipping Cost']].aggregate('sum').reset_index()
grouped_by_ship_mode_orders_df
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = (pct*total/100.0)/1000
        return '{v:.2f} K'.format(v=val)
    return my_autopct

values = grouped_by_ship_mode_orders_df['Shipping Cost']
labels = grouped_by_ship_mode_orders_df['Ship Mode']


plt.pie(values, labels= labels,autopct=make_autopct(values), wedgeprops=dict(width=0.5))



plt.title('Shipping cost by ship mode')
```

```python
sales_by_customer_df = pd.merge(orders_df, customers_df,  how='left', left_on=['Customer ID'], right_on = ['Customer ID'])[['Customer Name', 'Sales']]
sales_by_customer_df = sales_by_customer_df.groupby('Customer Name').aggregate('sum').sort_values('Sales')
sales_by_customer_df.head()
```

```python
ax = sales_by_customer_df[:20].plot(kind='barh', figsize=(15,20), zorder=1)
```

```python
orders_df['Delivery time'] = (orders_df['Ship Date'] - orders_df['Order Date']).dt.days
orders_df.head()
```

```python
avarage_delivery_time_orders_df = orders_df.groupby('Ship Mode')[['Ship Mode','Delivery time']].aggregate('mean').reset_index().sort_values('Delivery time')
avarage_delivery_time_orders_df
```

```python
f, ax = plt.subplots(figsize =(15,8))

ax.set_ylim(1.7,2)


plt.bar(avarage_delivery_time_orders_df['Ship Mode'],avarage_delivery_time_orders_df['Delivery time'])
```
